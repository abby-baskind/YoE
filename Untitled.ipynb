```python
df = dd.isel(y = slice(25,40), x = slice(100,120))
PH = df.ph
```

```python
d = xr.Dataset()
d = d.assign_coords({'y': PH.y, 'x': PH.x})
dat = np.zeros((len(d.y), len(d.x)))
dat[:] = np.nan
d['YOEpH'] = xr.Variable(dims = ['y','x'], data = dat)
d
```

```python
noise = 0.03
for i in d.y:
    for j in d.x:
        # dd.lat.values:
        a = PH.sel(x = j, y = i)
        px = np.poly1d(np.polyfit(a.t, a, 5))
        b = a.isel(t = 0).values.tolist() - 2*noise
        r = np.roots(px - b)
        r = r.real[abs(r.imag)<1e-5]
        r = r[r > 1850].min()
        # if a.size > 0:
        d["YOEpH"] = xr.where((d.coords["x"] == j) & (d.coords["y"] == i), r, d["YOEpH"])
```
